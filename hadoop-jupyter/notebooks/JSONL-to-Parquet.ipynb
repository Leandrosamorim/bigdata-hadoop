```python
import pandas as pd
from pathlib import Path
```

```python
bus_folder = Path.cwd().parent / "datasets" / "bus-api"
bus_parquet = Path.cwd().parent / "datasets" / "bus-api-parquet"
```

```python
dates = [
    "2023-05-15",
    "2023-05-16",
    "2023-05-17",
    "2023-05-18",
    "2023-05-19",
    "2023-05-20",
    "2023-05-21",
    "2023-05-22",
    "2023-05-23",
    "2023-05-24"
]
```

```python
i = 0
for dt in dates:
    for file in bus_folder.glob(f"{dt}_*.jsonl"):
        i += 1
        name = file.name.replace(':', '')
        if i < 78:
            print(f"{i} - skipping {name}.parquet")
            continue
        df = pd.read_json(file, lines=True)
        print(f"{i} - {name}.parquet")
        df.to_parquet(bus_parquet / f"{name}.parquet", engine="pyarrow")
```
